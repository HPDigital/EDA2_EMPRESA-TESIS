```python
df=pd.read_excel('C:/Users/HP/Desktop/TESIS/EMPRESA_NDA/DATOS/DATOS_NDA.xlsx', parse_dates=True)
df['Fecha']=pd.to_datetime(df['Fecha'])
df.head()
```
